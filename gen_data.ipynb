```python
def save_graph_indicator(num_graphs, num_nodes,path='1_100/'):
    with open(path+'AG_graph_indicator.txt', 'w') as file:
        for i in range(num_graphs):
            for j in range(num_nodes):
                file.write(str(i+1))
                file.write('\n')
```

##### Adjacency matrix

```python
def save_adjacency_matrix(num_graphs, gen_locus_, path='1_100/'):
    with open(path+'AG_A.txt', 'w') as file:
        for i in range(num_graphs):
            for k, v in gene_locus_int.items():
                if len(v) > 0:
                    combs = list(combinations(range(len(v)),2))
                    for i,j in combs:
                        line1 = '\t' + str(v[i]) + ',\t' + str(v[j])
                        file.write(line1)
                        file.write('\n')
                        line2 = '\t' + str(v[j]) + ',\t' + str(v[i])
                        file.write(line2)
                        file.write('\n')
```

```python
def get_adj_list(gene_locus_):
    adj_list = []
    for k, v in gene_locus_int.items():
        if len(v) > 0:
            combs = list(combinations(range(len(v)),2))
            for i,j in combs:
                adj_list.append((v[i],v[j]))
        # if k>5:
        #     break
    return adj_list
```

```python
get_adj_list(gene_locus_int)
```

```python
def get_graph(num_nodes, gene_locus_, node_attributes, graph_labels):
    '''
        graph_labels: GEX vector
    '''
    node_list = list(range(num_nodes))
    G=nx.Graph()
    G.add_nodes_from(node_list)
    G.add_edges_from(get_adj_list(gene_locus_))
    for u in G.nodes:
        G.nodes[u]['feat'] = node_attributes[u]
    
    G.graph['feat_dim'] = 1
    G.graph['label'] = graph_labels
    return G
```
